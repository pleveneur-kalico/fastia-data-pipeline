# Analyse des Biais - FastIA Module 2

Ce notebook analyse les biais potentiels dans le dataset nettoyé.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json

# Chargement des données traitées
df = pd.read_json('data/processed/dataset_fastia_clean_v1.jsonl', lines=True)
print(f"Dataset chargé : {len(df)} lignes")
```

## 1. Distribution des Catégories et Priorités

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='categorie', hue='priorite')
plt.title("Distribution des Catégories par Priorité")
plt.xticks(rotation=45)
plt.show()
```

## 2. Analyse de la Longueur du Texte

On vérifie si la longueur du texte (`input`) est corrélée à la catégorie.

```python
df['input_len'] = df['input'].str.len()

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='categorie', y='input_len')
plt.title("Longueur du texte par Catégorie")
plt.xticks(rotation=45)
plt.show()
```

## 3. Analyse des PII

Vérification du masquage des données sensibles.

```python
pii_tags = ['[EMAIL]', '[PHONE]', '[URL]', '[NAME]']
for tag in pii_tags:
    count = df['input'].str.contains(tag, regex=False).sum()
    print(f"Nombre de lignes contenant {tag} : {count}")
```
